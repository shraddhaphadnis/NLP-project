```python
import urllib
from bs4 import BeautifulSoup as bs
import nltk
import pandas  as pd
import numpy as np
from nltk.stem.snowball import SnowballStemmer
from nltk import ngrams
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from collections import Counter
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd
df = pd.read_csv('news_summary.csv')
```

```python
def clean_data(html,baseurl):
    soup = bs(html, 'html.parser')
    if baseurl == 'i':
        entries = soup.find_all('div',{'class':'description'})
    elif baseurl == 't':
        entries = soup.find_all('div',{'itemprop':'articleBody'})
    elif baseurl == 'h':
        entries = soup.find_all('div',{'itemprop':'articlebody'})
    for each in entries:
            if each.figure:
                each.figure.decompose()
    content = []
    for e in entries:
         content.extend(e.find_all("p"))
    
    text = ""
    for each in content:
        text = text + each.get_text() +" "

    text = text.encode('utf-8').decode("unicode_escape").encode('ascii','ignore')
#     text = nltk.sent_tokenize(text)    
    return text
```

```python
def readUrl(url):
    if "indiatoday" in url or "intoday" in url:
        baseurl = "i"
    elif "hindustantimes" in url:
        baseurl = "h"
    elif "theguardian" in url:
        baseurl = "t"

    file = urllib.urlopen(url)
    html = file.read()

    text = clean_data(html,baseurl)
    return text
```

```python
data = []
df1 = list(df['read_more'])
df1 = df1[:100]
len (df1)
for doc in df1:
    print doc
    data.append(readUrl(doc))
```

```python
sentences = []
for d in data:
    sen = d.split(". ")
    sentences.extend(sen)
text  = sentences
print text
```

```python
from gensim.models import Word2Vec
import numpy as np
from nltk.tokenize import RegexpTokenizer

sentences = []
for s in text:   
   # print s
    tokenizer = RegexpTokenizer(r'\w+')
    sentence = tokenizer.tokenize(str(s))
    sentences.append(sentence) 

# convert words to vectors
model = Word2Vec(sentences,
        size=150,
        window=5,
        min_count=1,
        workers=10)
print model
```
